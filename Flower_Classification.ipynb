```python
from fastai.vision.all import *
```

```python
df=!unzip /content/flowers.zip -d path_to_directory
```

```python
path=Path("/content/path_to_directory/flowers")
```

```python
data=ImageDataLoaders.from_folder(path, valid_pct=0.2, item_tfms=Resize(224))
```

```python
data.show_batch(figsize=(15,10))
```

```python
model=cnn_learner(data,resnet101, metrics=[error_rate, accuracy])
```

```python
model.lr_find()
```

```python
model.fine_tune(2, 0.01)
```

```python
model.show_results(figsize=(15,10))
```

```python
interp=Interpretation.from_learner(model)
interp.plot_top_losses(9,figsize=(20,10))
```

```python
model.save('final_model')
```
